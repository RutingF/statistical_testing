```python
import numpy as np
import pandas as pd 
from scipy import stats
from scipy.stats import ttest_ind 
```

### T-tests 
- Running Two-Sample T-Test below to compare means of two independent datasets 

```python
# Dummy data to demonstrate t-tests 

df1 = pd.DataFrame([np.random.random(100) for x in range(100)])
df2 = pd.DataFrame([np.random.random(100) for x in range(100)])
print(df1.shape, df2.shape)
df1.head()
```

```python
# for a given row inside of df1, is it the same as the row inside df2? 
# Critical value = 0.05, or alpha of 5% 
# goal: report when the p-value is less than 5% (meaning that we have sufficient evidence to say that columns are different)

def test_columns(alpha=0.05): 
    num_diff = 0 
    for col in df1.columns:
        # Run ttest between the 2 dataframes 
        teststat, pval = ttest_ind(df1[col], df2[col])
        # Check the pvalue versus the alpha
        if pval <= alpha: 
            # Print out if they're different and increment the num_diff
            print("col {} is statistically significantly different at alpha={}, pval={}".format(col, alpha, pval))
            num_diff = num_diff+1 

    # Print out summary stats 
    print("Total number different was {}, which is {}%".format(num_diff, float(num_diff)/len(df1.columns)*100))

# Call the function 
test_columns()
```
